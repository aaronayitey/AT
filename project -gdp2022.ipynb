```python
cat_columns_list=["Country Name","Country Code","Indicator Name","Indicator Code"]
new_column_names=[]
for column_name in gdp2022.columns:
    if column_name not in cat_columns_list:
        new_column_names.append(int(column_name))
    else:
        new_column_names.append(column_name)
print(new_column_names)
```
